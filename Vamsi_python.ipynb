```python
def frequency_sort_string(s):
    freq = {}
    # Count character frequency using loop
    for char in s:
        if char in freq:
            freq[char] += 1
        else:
            freq[char] = 1

    # Convert dict to list of tuples
    items = []
    for key in freq:
        items.append((key, freq[key]))

    # Sort by frequency (descending) then alphabetically
    for i in range(len(items)):
        for j in range(i + 1, len(items)):
            if (items[i][1] < items[j][1]) or (items[i][1] == items[j][1] and items[i][0] > items[j][0]):
                items[i], items[j] = items[j], items[i]

    # Build result string using loop
    result = ""
    for char, count in items:
        for _ in range(count):
            result += char

    # Capitalize first character
    result = result.capitalize()
    return result

# Main
def main():
    s = input("Enter the string: ")
    print(frequency_sort_string(s))

main()
```

```python
def are_complementary(s1, s2):
    combined = s1 + s2
    if len(combined) != 26:
        return "no"
    if set(combined) == set("ABCDEFGHIJKLMNOPQRSTUVWXYZ"):
        return "yes"
    else:
        return "no"
s1 = input()
s2 = input()
print(are_complementary(s1, s2))

```

```python
def run_length_encode(s):
    result = ""
    i = 0
    while i < len(s):
        count = 1
        while i+1 < len(s) and s[i] == s[i+1]:
            count += 1
            i += 1
        result += s[i].upper() + str(count)
        i += 1
    return result

# Main
s = input()
print(run_length_encode(s))
```
